```python

import glob
import os
import shutil

class_dict = {'0':'license_plate_white', '1':'license_plate_green', '2':'license_plate_blue', '3':'license_plate_yellow', '4':'license_plate_bike'}

def classification(images_path, class_num, out_path):
    flist = glob.glob(images_path)
    label_list = [lbf for lbf in flist if lbf.endswith(".txt")]
    label_list.sort()
    class_name = class_dict.get(class_num)
    out_path_ = out_path + '\\' + class_name
    createFolder(out_path_)
    for label_file_path in label_list:
        with open(label_file_path, 'r') as f:
            label_num = f.readline()[0]
            if class_num == label_num:
                shutil.copy(f.name, out_path_)
                print("copy labels")
                label_name = f.name.split('.')[0]
                shutil.copy(label_name + '.jpg', out_path_)
                print("copy images")
                print("------------------------------------------")
            elif label_num == "l":
                shutil.copy(f.name, out_path_)
                print("copy classes.txt")
        f.close()


def createFolder(dir):
    try:
        if not os.path.exists(dir):
            os.makedirs(dir)
    except OSError:
        print('Error: 해당 경로에 폴더를 만들 수 없음 ' + dir)
    return dir


def move_for_valid(images_path, out_path):
    flist = glob.glob(images_path)
    images_list = [imf for imf in flist if imf.endswith(".jpg")]
    images_list.sort()
    createFolder(out_path)
    for images_file_path in images_list:
        image_name = images_file_path.split('.')[0].split('\\')[-1]
        label_name = images_file_path.split('.')[0] + '.txt'
        if int(image_name) % 13 == 0:
            shutil.move(images_file_path, out_path + '\\' + image_name + '.jpg')
            print("move images")
            shutil.move(label_name, out_path + '\\' + label_name.split('\\')[-1])
            print("move labels")
            print("------------------------------------------")
            
            
def detections_more(images_path):
    flist = glob.glob(images_path)
    label_list = [lbf for lbf in flist if lbf.endswith(".txt")]
    label_list.sort()
    for label_file_path in label_list:
        with open(label_file_path, 'r') as f:
            labels = f.readlines()
            objs = len(labels)
            if objs > 1:
                print(objs)
                print(f.name)
```
